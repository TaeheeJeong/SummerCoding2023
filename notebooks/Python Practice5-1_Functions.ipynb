## Functions

built-in function

```python
print('Hello, World!')
```

defined function

```python
def multiplyByThree(val):
    return 3 * val

multiplyByThree(4)
```

```python
def multiply(val1, val2):
    return val1 * val2
```

```python
multiply(3, 4)
```

```python
a = [1,2,3]

def appendFour(myList):
    myList.append(4)
    
appendFour(a)
print(a)
```

```python
myList = [5,4,3,2]
sorted(myList)
```
